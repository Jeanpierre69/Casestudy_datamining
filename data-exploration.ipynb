# Train dataset verkenning

WARNING: NIET TEVEEL ROWS LEZEN ANDERS RAM NAAR DE ZAK

```python
import polars as pl

dataPath = "data/"
```

```python
# aantal rows die je van parquet file wil lezen om memory safe te blijven
def preview(path, rows):
    df = pl.read_parquet(path, n_rows=rows)
    print(df.shape)
    print(df.columns)
    print(df.head(10))
```

```python
from pathlib import Path
from time import sleep

TRAIN_DIR = Path("data/train")
files = sorted(TRAIN_DIR.glob("*.parquet"))
print(f"Found {len(files)} CSV files in {TRAIN_DIR}")

for t in files:
    sleep(1)
    print('\n' + '='*80)
    print(f"Preview for: {t.name}")
    try:
        preview(t, 10)
    except Exception as e:
        print(f"Error reading {t.name}: {e}")
print('\nDone')
```
